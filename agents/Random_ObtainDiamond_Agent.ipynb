```python
import random
import numpy as np
import torch as th
from torch import nn
import gym
import minerl
from tqdm.notebook import tqdm
from colabgymrender.recorder import Recorder
from pyvirtualdisplay import Display
from sklearn.cluster import KMeans
import logging
logging.disable(logging.ERROR) # reduce clutter, remove if something doesn't work to see the error logs.
from sklearn.model_selection import train_test_split
from time import time
```

## Random baseline

```python
minerl.data.download(directory='data', environment='MineRLObtainDiamond-v0');
```

```python
env = gym.make('MineRLObtainDiamond-v0')
# env = Recorder(env, './video', fps=60)
episodes = 3
stats = {'runtime': [], 'reward': []}

for episode in range(1, episodes + 1):
    start = time()
    obs = env.reset()
    done = False
    score = 0

    while not done:
        # Take a random action
        action = env.action_space.sample()
        # In BASALT environments, sending ESC action will end the episode
        # Lets not do that
        # action["ESC"] = 0
        obs, reward, done, _ = env.step(action)
        score += reward
        #emv.render()
        #env.render(mode = 'rgb_array')
        if done:
            print("\nobtained diamond\n")
        # env.render(mode = 'human')
    
    stats['runtime'].append(time() - start)
    stats['reward'].append(score)
```

```python
import json
with open('stats_random.json', 'w') as outfile:
    json.dump(stats, outfile)
```
